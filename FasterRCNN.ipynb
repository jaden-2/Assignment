## Problem 4: Create a file for learning

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
```

```python

annot = pd.read_csv("/content/drive/MyDrive/ObjectDetection/annotation.txt")
```

```python
annot.head()
```

```python

annot.iloc[:, 0] = annot.iloc[:, 0].apply(lambda x: "/"+ x)
annot.iloc[:, 5] = encoder.fit_transform(annot.iloc[:, 5])
```

```python
annot.tail()
```

```python
output = "annotation.txt"

with open(output, "w") as f:
  f.write(annot.to_string(index=False, header=False))
```
