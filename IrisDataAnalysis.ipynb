### Problem 6

```python
#Make a pie chart of the number of samples per label (while showing percentages)

species_info = df.groupby(["Species"]).count()
labels = data.target_names

for key in species_info.keys():
  fig, ax = plt.subplots(figsize=(5, 5))
  ax.set_title(key)
  ax.pie(species_info[key], autopct= lambda pct: f"{pct :.1f}%")
  fig.legend(labels)
plt.show()
```

```python
#Select features one by one and visualize the distribution of data for each label using a box plot
features = df.keys()
species = data.target_names
for feature in features[:-1]:
  feat = []
  for i in range(species.size):
    feat.append(df[df['Species']==i][feature])
  fig, ax = plt.subplots(figsize=(7, 7))
  ax.boxplot(feat)
  ax.set_title(feature)
  ax.grid()
  fig.legend(species)
plt.show()
```

```python
#Select features one by one and visualize the distribution of data for each label using a violin plot
features = df.keys()
species = data.target_names
for feature in features[:-1]:
  feat = []
  for i in range(species.size):
    feat.append(df[df['Species']==i][feature])
  fig, ax = plt.subplots(figsize=(7, 7))
  ax.violinplot(feat)
  ax.set_title(feature)
  ax.grid()
  fig.legend(species)
plt.show()
```

### Problem 7

```python
#

#get a list of all columns excluding the last
features = df.keys()[:-1]

#get a list of the species of flowers available
species = data.target_names

#for each feature, group by species
grouped_iris = {}

for feature in features:
  grouped_iris[feature] = []
  for specie in range(species.size):
    grouped_iris[feature].append(df[df["Species"] == specie][feature])
```

```python
#Make scatter plots of all combinations (six), such as choosing two from four features, color-coded by type.
for x in range(4):
  for i in range(4):
    if x == i:
      continue
    #print((x, (x-i) % 4))
index = 0
keys = list(grouped_iris.keys())

for i in range(len(keys)):
  x, y = "", ""
  for a in range(len(keys)):
    if i == a:
      continue
    x = keys[i]
    y = keys[a]
    fig = plt.figure(figsize=(10, 6))
    ax = fig.add_subplot(111)
    ax.scatter(grouped_iris[x][0], grouped_iris[y][0], c="r")
    ax.scatter(grouped_iris[x][1], grouped_iris[y][1], c="b")
    ax.scatter(grouped_iris[x][2], grouped_iris[y][2], c="y")
    ax.set_title(f"A graph of {x} VS {y}")
    plt.xlabel(x)
    plt.ylabel(y)
    ax.legend(species)
    ax.grid(True)


```
