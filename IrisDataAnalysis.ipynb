```python
import numpy as np
from sklearn.datasets import load_iris
import pandas as pd
import matplotlib.pyplot as plt
```

### Problem 1

```python
data = load_iris(as_frame=True)

x = pd.DataFrame(data["data"])

y = pd.DataFrame(data["target"])

y.columns= ["Species"]

print(x.head(), y.head(),sep="\n")

```

### Problem 2

```python
df = pd.concat([x, y], axis=1)

df.head()
```

### Problem 6

```python
#Make a pie chart of the number of samples per label (while showing percentages)
z = df.count()

def func(pct):
  return f"{pct:.1f}%"

fig, ax = plt.subplots(figsize= (10, 5))
ax.pie(z, autopct= lambda pct: func(pct), labels=df.keys())

ax.set_title("Pie chart of the number of samples per label")
plt.show()
```

```python
#Select features one by one and visualize the distribution of data for each label using a box plot
```

```python
#Select features one by one and visualize the distribution of data for each label using a violin plot

for key in df.keys():
  plt.title(key)
  plt.violinplot(df[key])
  plt.show()
  print()
```
