```python
import matplotlib.pyplot as plt


grainsPerDay = []

for x in range (100):
    numOfGrains = 1 * (2**x) #number of grains of rice for each day
    grainsPerDay.append(numOfGrains)

plt.title("Sorori Problem")
plt.xlabel("Number of days")
plt.ylabel("Number of grains")
plt.tick_params(labelsize=20) # Make settings related to axis values
plt.plot(grainsPerDay)
plt.show()

#print(grainsPerDay)
```

```python
def sorori_prob(numOfDays =100):
    grainsPerDay = []
    t_grainsPerDay = []
    for x in range (numOfDays):
        numOfGrains = 1 * (2**x) #number of grains of rice for each day
        grainsPerDay.append(numOfGrains)
        total = sum(grainsPerDay[:x]) #sum of rice of all the previous days till current day
        t_grainsPerDay.append(total)
    
    return grainsPerDay, t_grainsPerDay

list_n_grains, total_n_grains = sorori_prob(100)
```

```python
"""
According to google, an average person can eat 1/2 cup of rice
1 cup of rice contains approx 5000 grains of rice
"""

def get_rice_reserve(numOfGrains):
    """
    Calculate the number of days 50 students woud survive on rice reserve
    """
    numOfStudents = 50
    t_num_of_rice_required = numOfStudents * 2500 #total number of rice 50 students would eat in a day
    numOfDaysToLive = int(numOfGrains / t_num_of_rice_required)
    return numOfDaysToLive


for grains in total_n_grains[::5]:
    print(f" 50 Students will live for {get_rice_reserve(grains)} days on {grains} grains of rice")
```
