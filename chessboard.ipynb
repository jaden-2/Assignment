```python
import numpy as np
import matplotlib.pyplot as plt
```

## Problem 1

```python
#create a numpy array containing 0 - 3 to represent a board
small_board_ndarray = np.arange(4) 

#doubling the wheat added to each cell in the board
small_board_ndarray = 2 ** small_board_ndarray

#reshape dimension to 2 by 2 array
small_board_ndarray = small_board_ndarray.reshape((2, 2)).copy()

print(f"Arrange wheat on a plate of 4 squares (ndarray)：\n{small_board_ndarray}")

```

## Problem 2

```python
def expand(a, b):
    ndarry = np.arange(a*b, dtype=np.uint64)
    
    ndarry = 2 ** ndarry
    return ndarry.reshape((a, b)).copy()


chessboard = expand(8, 8)

print(chessboard)
```

## Problem 3

```python
total_wheat = expand(8, 8).sum()
print(f"Total number of wheat on a chess board of 8 x 8 squares: {total_wheat :,}")


chessboard = expand(8, 8)
column = np.arange(1, 9)
number = np.hsplit(chessboard, 8)

for x in range(len(number)):
    number[x] = number[x].sum()

plt.xlabel("column")
plt.ylabel("number")
plt.title("number in each column")
plt.bar(column, number)
plt.show()
```

## Problem 4

```python
plt.xlabel("column")
plt.ylabel("row")
plt.title("heatmap")
plt.pcolor(chessboard)
plt.show()
```
