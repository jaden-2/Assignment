```python
import numpy as np
import matplotlib.pyplot as plt
```

### Problem 1

```python
mean = (-3, 0)
cov = np.array([[1, .8], [.8, 1]])

rand = np.random.multivariate_normal(mean, cov, (500,))

np.random.seed(0)
print(rand.shape)
```

### Problem 2

```python
x, y = np.hsplit(rand, 2)

plt.scatter(x, y)

plt.xlabel("X")
plt.ylabel("Y")
plt.title("Scatter plot of random data")
plt.show()
```

### Problem 3

```python
plt.hist(x, 50)

plt.xlim((x.min(), x.max()))

plt.xlabel("X")
plt.ylabel("Frequency")
plt.title("Histogram")
plt.show()
```

```python
plt.hist(y, 50)

plt.xlim((y.min(), y.max()))

plt.xlabel("Y")
plt.ylabel("Frequency")
plt.title("Histogram")
plt.show()
```

### Problem 4

```python
mean = (0, -3)
cov = np.array([[1, .8], [.8, 1]])

rand1 = np.random.multivariate_normal(mean, cov, 500)

np.random.seed(1)

x1, y1 = np.hsplit(rand1, 2)

plt.scatter(x, y)

plt.scatter(x1, y1, color="r")
plt.legend([0, 1])

plt.xlabel("X")
plt.ylabel("Y")
plt.title("Scatter plot")
plt.show()
```

### Problem 5

```python
bundle = np.vstack((rand, rand1))

bundle.shape
```

### Problem 6

```python
""" Since the size of both random arrays generated are equal
I can create a row array of ones and change the fist 500 elements
After which, I would concatenate on the horizontal axis"""

sample = np.ones((1000, 1))
sample[:500] = 0
labeled = np.hstack((bundle, sample))

print(labeled.shape) #sanpling of the 1000 rows of data stored

```
